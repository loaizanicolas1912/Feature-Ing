```python
import pandas as pd

datos=pd.read_csv("AB_NYC_2019.csv")
df=pd.DataFrame(datos)
df
```

```python
df.head()
```

```python
#OBSERVAMOS SUS ESTADISTICAS
df["price"].describe()
#PODEMOS OBSERVAR SUS LIMITES, EN ESTE CASO TOMARE VALORES ATIPICOS PARA LAS QUE ESTAN DEBAJO DEL PORCENTIL MINIMO DE 5% 
#Y VALORES QUE ESTEN POR ENCIMA DEL PORCENTIL DE 97%
```

```python
#VAMOS A ESTUDIAR LOS VALORES ATIPICOS QUE TIENE DICHA COLUMNA PARA ESTO REALIZAMOS EL CALCULO PARA UN PORCENTIL DEL 95%
#LOS VALORES QUE SE ENCUENTREN POR ENCIMA DE ESTOS SERAN VALORES ATIPICOS
Max_cuartil=df["price"].quantile(0.97)  #NO SIEMPRE ES 95% ESTO PUEDE DEPENDER DE UNA PAUTA QUE SE DE
Max_cuartil
#GUARDAREMOS ESTE DATO EN UNA VARIABLE
```

```python
#AHORA HAREMOS LO MISMO CON EL MINIMO PARA CREAR EL UMBRAL
Min_cuartil=df["price"].quantile(0.05)
Min_cuartil
#YA SABEMOS QUE LO QUE SE ENCUENTRE POR DEBAJO DE ESTE ES UN VALOR ATIPICO 
```

```python
#AHORA VAMOS A IDENFICICAR LOS VALORES QUE SON ATIPICOS DE UNA MANERA VISUAL
df[df["price"]>Max_cuartil]
```

```python
#AL IGUAL PARA LOS ATIPICOS MENORES
df[df["price"]<Min_cuartil]
```

```python
#UNA MENERA MAS CORTA DE REALIZAR EL CODIGO ES 
Min_cuartil2, Max_cuartil2 = df["price"].quantile([0.05,0.97])
Min_cuartil2, Max_cuartil2
```

```python
#AHORA VAMOS A SELECCIONAR LOS DATOS QUE SE ENCUENTRE EN ESTE UMBRAL, Y SE GUARDARAN EN UN NUEVO DATAFRAME
df2 = df[(df["price"]>Min_cuartil2) & (df["price"]<Max_cuartil2)]
df2
```
